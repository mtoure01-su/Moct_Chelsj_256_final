```python
# Start of function definitions
def enoughSentences(sen_no):
    if sen_no >= 5:
        return(True)
    else:
        return(False)
def nottoomanySentences(sen_no):
    if sen_no <= 10:
        return(True)
    else:
        return(False)

def enoughPages(count, page_min):
    if type(page_min) == int and type(count) == int:
        if count >= int(page_min):
            return(True)
        else:
            return(False)

def nottoomanyPages(count, page_max):
    if type(page_max) == int and type(count) == int:
        if count <= int(page_max):
            return(True)
        else:
            return(False)

def correctConclusion(conclusion):
    conclusion = conclusion.lower()
    bad_conclusions = ["to conclude", "in conclusion", "finally", "to summarize", "to sum up", "in summary"]
    good_conclusions = ["given", "in the final analysis", "on the whole", "thus"]
    good_string = ''
    good_con_true = False
    bad_con_true = False
    
    for string in good_conclusions:
        if string == good_conclusions[-1]:
            good_string = good_string+' or '+string
        elif string == good_conclusions[0]:
            good_string = string
        else:
            good_string = good_string+', '+string
        
    for phrase in bad_conclusions:
        conclusion_pos = conclusion.find(phrase, 0, len(conclusion))
        if conclusion_pos == 0:
            bad_con_true = True
            print('You have started your conclusion with "%s" and it is generally advised to stay away from redundant transitional words. Might I suggest replacing "%s" with %s?' % (phrase, phrase, good_string))
    
    for phrase in good_conclusions:
        conclusion_pos = conclusion.find(phrase, 0, len(conclusion))
        if conclusion_pos == 0:
            good_con_true = True
            print('You have started your conclusion with "%s". Great transitional word!' % (phrase, phrase, good_string))
            
    if good_con_true == False and bad_con_true == False:
        print('No clear problems in your conclusion. Might I suggest some great transitional words to begin your conclusions: %s' % (good_string))
            
#Main Code

try:
    ## Creating the Document
    with open('essay.txt', 'w') as f_write:
        f_write.write("")


    page_min = int(input("What is the minimum amount of pages required of you? "))
    page_max = int(input("What is the maximum amount of pages required of you? "))

    ##Inputing the Paragraphs
    while True:
        para = input("Enter a paragraph or type 'done': ")

        if para != "done":
            para = str(para)
            with open('essay.txt', 'a') as f_append:
                f_append.write(para+"\n") 
        else:
            break
    ##Reading/splitting the text
    with open('essay.txt', 'r') as f_read:
        entire_text = ''
        sen_no_list = []
        for line in f_read.readlines():
            line = line.strip()
            para_sentence = line.split('. ')
            line = line.strip()
            entire_text = entire_text+' '+line
            para_sen_no = len(para_sentence)
            sen_no_list.append(para_sen_no)

        entire_text = entire_text.strip()
        each_char = list(entire_text)
        char_no = len(each_char)

        each_word = entire_text.split(' ')
        word_no = len(each_word)

    print('\n'+'General Document Info: ')
    print('The number of sentences is %i' % (para_sen_no) )
    print('The number of characters is %i' % (char_no) )
    print('The number of words is %i' % (word_no) )

    ## Checking parameters of text
    ### For paragraph sentence check
    print('\n'+'Paragraph Sentence Check: ')
    para_no = 1

    for sen_no in sen_no_list:
        enough = enoughSentences(sen_no)
        too_many = nottoomanySentences(sen_no)
        sen_add = 5 - sen_no
        sen_remove = sen_no - 10
        if enough == False:
            enough_pos = para_no
            print('Paragraph %i is too short. It needs at least %i more sentence(s).' % (enough_pos, sen_add))
        elif too_many == False:
            too_many_pos = para_no
            print('Paragraph %i is too long. Cut it down by at least %i sentence(s).' % (too_many_pos, sen_remove))
        elif enough == True and too_many == True:
            print('Paragraph %i seems to have the appropriate amount of sentences. Great job!' % (para_no))
        para_no = para_no + 1

    ### For page number check
    page_count = math.ceil((word_no / 230))

    print('\n'+'Page Number Check: ')
    enough = enoughPages(page_count, page_min)
    too_many = nottoomanyPages(page_count, page_max)
    pages_add = page_min - page_count
    pages_remove = page_count - page_max
    if enough == False:
        print('Your paper is %i page(s) and is too short. Add at least %i page(s).' % (page_count, pages_add))
    elif too_many == False:
        print('Your paper is %i page(s) and is too long. Cut it down by at least %i page(s).' % (page_count, pages_remove))
    elif enough == True and too_many == True:
        print('Your paper seems to be the appropriate amount of pages. Great job!')

    ### Contractions
    print('\n'+'Contractions Check: ')
    contraction_dict = {"ain't": "are not", "aren't": "are not", "can't": "can not", "cain't": "can not", 
                        "could've": "could have", "couldn't": "could not", "could of": "could have", 
                        "doesn't": "does not", "don't": "do not", "gonna": "going to", "gotta": "have to", 
                        "hadn't": "had not", "hasn't": "has not", "haven't": "have not", "he'd": "he had",
                        "he'll": "he will", "he's": "he is", "how'd": "how would", "how'll": "how will",
                        "how's": "how is", "I'd": "I would", "I'll": "I will", "I'm": "I am", "I've": "I have", 
                        "isn't": "is not", "it'd": "it would", "it'll": "it will", "it's": "it is", "let's": "let us",
                        "mayn't": "may not",  "may've": "may have", "mightn't": "might not","might've": "might have",
                        "mustn't": "must not", "must've": "must have", "needn't": "need not", "ol'": "old", 
                        "oughtn't": "ought not", "shan't": "shall not", "she'd": "she would", "she'll": "she will",
                        "she's": "she is", "should've": "should have", "shouldn't": "should not", "somebody's": "somebody is", 
                        "someone's": "someone is", "something's": "something is", "that'll": "that will", 
                        "that're": "that are", "that's": "that is", "that'd": "that would", "there'd": "there would",
                        "there're": "there are", "there's": "there is", "these're": "these are", "they'd": "they would",
                        "they'll": "they will", "they're": "they are", "they've": "they have", "this's": "this is", 
                        "those're": "those are", "'tis": "it is", "'twas": "it was", "'twasn't": "it was not", "wasn't": "was not",
                        "we'd": "we would", "we'd've": "we would have", "we'll": "we will", "we're": "we are",
                        "we've": "we have", "weren't": "were not", "what'd": "what would", "what'll": "what will",
                        "what're": "what are", "what's": "what is", "what've": "what have", "when's": "when is", 
                        "where'd": "where did", "where're": "where are", "where's": "where is", "where've": "where have",
                        "which's": "which is", "who'd": "who did", "who'd've": "who would have", "who'll": "who will",
                        "who're": "who are", "who's": "who is", "who've": "who have", "why'd": "why did",
                        "why're": "why are", "why's": "why is", "won't": "will not", "would've": "would have",
                        "wouldn't": "would not", "y'all": "you all", "you'd": "you did", "you'll": "you will", 
                        "you're": "you are", "you've": "you have"}
    contraction_list = []
    contraction_dict_keys = contraction_dict.keys()

    for contraction in contraction_dict_keys:
        contraction_list.append(contraction)

    a_contract = False

    for word in each_word:
        for key in contraction_list:
            if word == key:
                de_contract = contraction_dict[word]
                a_contract = True
                print ('You have the contraction "%s". You might want to replace it with %s.' % (word, de_contract))
    if a_contract == False:
        print('You have no contractions in your paper. Great!')

    ### Conclusion Analysis
    print('\n'+'Conclusion Check: ')
    para_list = []
    para_no = 1
    with open('essay.txt', 'r') as f_read:
        for line in f_read.readlines():
            para_list.append(line)

    correctConclusion(para_list[-1])

    ### Spell Check
    print('\n'+'Spell Check: ')
    from textblob import Word
    for orig_word in each_word:
        checked_word = Word(orig_word)
        full_word_check = checked_word.spellcheck()
        for check in full_word_check:
            if check == full_word_check[0]:
                for check_word in check:
                    if check_word == check[0] and orig_word != check_word:
                        print('You wrote "%s", did you mean "%s" instead?' % (orig_word, check_word))
except ValueError:
    print('You entered letters/words here, you need to enter a number!' % ())
```
